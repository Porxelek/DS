```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis") # yapacagimiz task ismi
# yamlis yazdiginda hatada duzgununu verir duzeltirsin
classifier("I've been waiting for a HuggingFace course my whole life.")
# tahmin etmesi gereken cumle
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
    # her cumle virgul ile ayrilacak , liste olcak ve string olcak tmm cnmmm
)
```

```python
# pozitifi ile negatifi nasil cekcem 
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"])[0]["label"]
```

```python
y_pred = []
for i in ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]:
  # gelen yorumu aldim , calssifier icine i vercem listenin icine gircem ve dict olustuuryoruym ve labella pozitifi cekiyorum
  if classifier(i)[0]["label"] == "POSITIVE":
    y_pred.append(0)# pozitive sifir dediysen y predde ytestte 0 vermen lazim bu 0 sor ???
  else:
    y_pred.append(1)
y_pred
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
# cumle veriyon herhangi bir calss icin egitim yok asagidaki 3 classa gore egitim olcak
# istedigin kadar arttir azalt
# burda labellari belirlemek bize bagli
# cosinus similarty e bakarak tahmin etcek
# wrod token benzerliklerine bakarak karar veriyor kumeleme gibi 
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
classifier(
    ["world cup will be held in qatar this year", "ballet and theater are indispensable for me.", "This year, a new law on animal rights was enacted."],
    candidate_labels=["education", "politics", "business", "art", "sport", "justice","animal"],
)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
# bak ner modeli verdigin text icinde modelin ne uzerinde egitildiyse
# organizsayon , locasyon kisi, personel egittiysen o seklde egitilmi modelle calisman lazim

# o modelle preduction yapmalisin
# personel egittigine mekan soramazsin n prediction yapamazsin onla ilgili
# modelin egitimine gore tokenlerimizi clasify ediyor.
# ner modellerinin datasi ozel senin yapman cok zor cunku data sekli farkli
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```
